## Problem 1: More water needed
Let's say we decided that the system from the Begay home notebook doesn't provide enough flow. We want to use the same exact system, but triple the pipe diameter, so we can get a higher flow rate. Is this a reasonable solution? Why or why not?

```python
system = ps.PipeSystem('PVC',3*0.0127,1) # 0.0127 meters is about 0.5 inches
system.addSection(2,0,'Entrance')
system.addSection(1.5,1.5,'90-Degree Elbow')
system.addSection(0.5,0),'90-Degree Elbow'
system.addSection(0.2,0.2,'90-Degree Elbow')

system.addPump(4,sf.pcurve,0.000189271)

system.solve()
```

```python
system.plotCurves()
```

```python
system.plotAllHead()
```

```python
print('v = ' + str(system.v) + ' m/s')
```
